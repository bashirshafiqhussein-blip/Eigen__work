```python
import numpy as np
```

```python
B=np.array([[3,1],[0,2]])
```

```python
print("Matrix B:")
print(B)
```

```python
eigenvalues,eigenvectors=np.linalg.eig(B)
```

```python
print("\Eigenvectors:")
print(eigenvectors)
```

```python
# verify the result (B * V = B * V)
```

```python
for i in range(len(eigenvalues)):
    Bv = np.dot(B, eigenvectors[:, i])
    lv = eigenvalues[i] * eigenvectors[:, i]
    print(f"\nverification for Eigenvalue {eigenvalues[i]}:")
    print("B * V =",Bv)
    print("lv * v =",lv)  
```
